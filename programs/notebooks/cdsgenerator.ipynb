# Instalación de la TensorFlow Object Detection API

La siguiente celda contiene un script que automatiza totalmente la instalación de la API de detección de objetos de TensorFlow, descargándola a través de internet. Al finalizar el proceso, un script concreto de la API ubicado en la carpeta *scripts* del proyecto *ObjectRecognition* se importará, este será utilizado más adelante para exportar los datasets en el formato de TensorFlow.

```python
#@title
from requests import get
from zipfile import ZipFile
from os import chdir, remove
from shutil import copy, move, rmtree

# Directorio de trabajo base.
chdir(wDirectory)

# Crea un fichero ZIP en el disco.
with open('models.zip', 'wb') as file:

  # Descarga el contenido del ZIP y lo guarda en el fichero.
  file.write(get('https://github.com/tensorflow/models/archive/refs/heads/master.zip').content)

# Abre el ZIP en modo lectura.
with ZipFile('models.zip', 'r') as zip:

  # Extrae el contenido del ZIP en el directorio actual.
  zip.extractall('.')

# Renombra la carpeta descomprimida.
move('models-master', 'models')

# Elimina el fichero comprimido.
remove('models.zip')

# Cambia el directorio actual.
chdir('models/research')

# Compilación de los protobufs.
%sx protoc object_detection/protos/*.proto --python_out=.

# Copia el instalador de la TensorFlow Object Detection API en el directorio actual.
copy('object_detection/packages/tf2/setup.py', '.')

# Instalación de la TensorFlow Object Detection API.
%pip -q install .

# Cambia el directorio actual.
chdir('../..')

# Borra la carpeta descargada.
rmtree('models')

# Copia el script para generar los tf-records.
copy(join(pathToDriveTfOdaScripts, 'create_coco_tf_records.py'), '.')
```

# Coco Dataset Generator

Esta penúltima celda contiene un programa para generar utomáticamente *datasets* de imágenes basados en COCO, los cuales se podrán utilizar más adelante para el entrenamiento y validación de modelos de detector de la *TF Object Detection API*.

\\
Instrucciones para el uso del programa:

  1. Seleccionar las categorías que deseamos que contenga nuestro *dataset*.
  
  2. Para cada selección anterior, elegir la cantidad de imágenes de entrenamiento y validación que habrá.
  
  3. Pulsar el botón *Create Dataset*, esto hará que *Fiftyone* descargue de internet imágenes aleatorias de las categorías escogidas.
    * El tiempo en que tarde en completarse está acción es completamente dependiente de la velocidad de conexión a internet.
    * Cuando esta acción finaliza pueden visualizarse los conjuntos e entrenamiento y validación en la última celda.
    * Los conjuntos de entrenamiento y validación prevalecen en la BDD de *Fiftyone* hasta que son exportados a *Google Drive*. 
  
  4. Indicar un nombre para exportar el *dataset*.

  5. Pulsar el botón *Export To Google Drive*, esto provocará la exportacion a la carpeta *datasets*, en *ObjectRecognition*, en nuestro *Google Drive*, de una carpeta con el nombre indicado en (4) que contendrá:
    * Una carpeta con las imágenes del conjunto de entrenamiento, y otra con las del conjunto de validación.
    * Un fichero en el formato (\*.json) de COCO, que describe el conjunto de entrenamiento, y otro que describe el conjunto de validación.
    * Un fichero en el formato (\*.records) de TF, que describe el conjunto de entrenamiento, y otro que describe el conjunto de validación.
    * Un fichero de *summary* en formato (\*.json), que resume el contenido del *dataset*.
    * Un fichero de *label map*, que asigna identificadores enteros a los nombres de las clases.

```python
#@title
from json import dump
from logging import ERROR
from datetime import datetime
from shutil import move, rmtree
from os.path import join, isdir
from os import listdir, makedirs
from fiftyone.zoo import load_zoo_dataset
from fiftyone import Dataset, ViewField as F
from fiftyone.core.utils import LoggingLevel
from fiftyone.types import COCODetectionDataset
from fiftyone import load_dataset, list_datasets
from ipywidgets import Box, Text, Label, Button, Layout
from fiftyone.core.dataset import delete_datasets, dataset_exists
from ipywidgets import Checkbox, BoundedIntText, BoundedFloatText

# Máxima longitud para el nombre de los dataset.
DATASET_NAME_MAX_LENGTH = 80

# Clases definidas en el dataset COCO.
COCO_CLASS_NAMES = (
  'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus', 'train', 
  'truck', 'boat', 'traffic light', 'fire hydrant', 'stop sign', 
  'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow', 
  'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella', 'handbag', 
  'tie', 'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball', 'kite', 
  'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 
  'tennis racket', 'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 
  'bowl', 'banana', 'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 
  'hot dog', 'pizza', 'donut', 'cake', 'chair', 'couch', 'potted plant', 
  'bed', 'dining table', 'toilet', 'tv', 'laptop', 'mouse', 'remote', 
  'keyboard', 'cell phone', 'microwave', 'oven', 'toaster', 'sink', 
  'refrigerator', 'book', 'clock', 'vase', 'scissors', 'teddy bear', 
  'hair drier', 'toothbrush'
)

# Estilo común para los widgets.
COMMON_STYLE = {'description_width': 'initial'}

# Estilo para los widgets de tipo botón.
BUTTON_STYLE = {'font_weight': 'bold', 'button_color': 'lightgray'}

# Layout común para los widgets.
COMMON_LAYOUT = Layout(width='94%', margin='8px 8px 8px 8px')

# Configuración de los botones relacionados con el conjunto "training".
TRAINING_IMAGES_DEFAULT = 80
TRAINING_IMAGES_MIN = 80
TRAINING_IMAGES_MAX = 16000
TRAINING_IMAGES_STEP = 1

# Configuración de los botones relacionados con el conjunto "validation".
VALIDATION_IMAGES_DEFAULT = 20
VALIDATION_IMAGES_MIN = 5
VALIDATION_IMAGES_MAX = 2600
VALIDATION_IMAGES_STEP = 1

# Filas de Widgets.
widgetRows = []

# Widgets del FlexBox.
widgetList = []

# Se crea una fila de widgets (class, train, val) por cada clase de COCO.
for className in COCO_CLASS_NAMES:

  # Fila del FlexBox.
  widgetRow = []

  # CheckBox para seleccionar una clase concreta.
  chckbxClass = Checkbox(
      value=False, 
      description=className,
      style=COMMON_STYLE,
      layout=COMMON_LAYOUT
  )

  # Se añade el widget a la fila.
  widgetRow.append(chckbxClass)

  # Selector del número de imágenes del conjunto "training".
  iBndTxtTraining = BoundedIntText(
      value=TRAINING_IMAGES_DEFAULT,
      min=TRAINING_IMAGES_MIN,
      max=TRAINING_IMAGES_MAX,
      step=TRAINING_IMAGES_STEP,
      description='Training Images:',
      style=COMMON_STYLE,
      layout=COMMON_LAYOUT
  )

  # Se añade el widget a la fila.
  widgetRow.append(iBndTxtTraining)

  # Selector del número de imágenes del conjunto "validation".
  iBndTxtValidation = BoundedIntText(
      value=VALIDATION_IMAGES_DEFAULT,
      min=VALIDATION_IMAGES_MIN,
      max=VALIDATION_IMAGES_MAX,
      step=VALIDATION_IMAGES_STEP,
      description='Validation Images:',
      style=COMMON_STYLE,
      layout=COMMON_LAYOUT
  )

  # Se añade el widget a la fila.
  widgetRow.append(iBndTxtValidation)
  
  # FlexBox para mostrar los paneles horizontales.
  fbRowPanel = Box(
      children=widgetRow,
      layout=Layout(
          width='94%',
          display='flex',
          flex_flow='row',
          align_items='center',
          border='1px solid grey', 
          margin='10px 0px 10px 0px', 
          padding='10px 10px 10px 10px'
      )
  )

  # Se guardan las referencias a las filas de widgets.
  widgetRows.append(widgetRow)

  # Se añade la fila de widgets en un panel horizontal que los agrupa.
  widgetList.append(fbRowPanel)

# Caja de texto para escribir el nombre con el que exportar el dataset.
txtDatasetName = Text(
  description='Dataset Name:',
  value='',
  placeholder='Write here your dataset name',
  style=COMMON_STYLE,
  layout=COMMON_LAYOUT
)

# Función para gestionar los eventos de la caja de texto.
def txtDatasetNameHandler(change):
  
  # Comprobamos que el cambio sea el texto que se escribe.
  if change['name'] == 'value':
    
    # Se captura el valor actual de la caja de texto.
    value = change['new']

    # Índice que representa la posición donde se encuentra el primer espacio.
    index = value.find(' ')

    # Si el espacio está presente, se desprecia la mitad derecha de la cadena.
    if index != -1:
      txtDatasetName.value = value[0:index]

    # Si la longitud de texto sobrepasa cierto número de caracteres, se acorta.
    if len(value) > DATASET_NAME_MAX_LENGTH:
      txtDatasetName.value = value[0:DATASET_NAME_MAX_LENGTH]

# Se regista el gestor de eventos de la caja de texto.
txtDatasetName.observe(txtDatasetNameHandler)

# Se añade el widget a la lista de widgets del FlexBox.
widgetList.append(txtDatasetName)

# Botón para que el usuario pueda marcar todas las clases disponibles.
btnCheckAll = Button(
    description='Check All',
    style=BUTTON_STYLE,
    layout=COMMON_LAYOUT
)

# Función para establecer el estado de todos los CheckBox.
def setCheckBoxsTo(boolean):

  # Se recorren todas las filas de widgets.
  for widgetRow in widgetRows:

    # Establece el valor del CheckBox al pasado por parámetro.
    widgetRow[0].value = boolean

# Manejador del evento OnClick.
btnCheckAll.on_click(lambda obj : setCheckBoxsTo(True))

# Se añade el widget a la lista de widgets del FlexBox.
widgetList.append(btnCheckAll)

# Botón para que el usuario desmarque todos los CheckBox.
btnUncheckAll = Button(
    description='Uncheck All',
    style=BUTTON_STYLE,
    layout=COMMON_LAYOUT
)

# Manejador del evento OnClick.
btnUncheckAll.on_click(lambda obj : setCheckBoxsTo(False))

# Se añade el widget a la lista de widgets del FlexBox.
widgetList.append(btnUncheckAll)

# Botón para la creación de datasets.
btnCreateDataset = Button(
    description='Create Dataset',
    style=BUTTON_STYLE,
    layout=COMMON_LAYOUT
)

# Clases seleccionadas por el usuario.
selectedClasses = []

# Función para crear un dataset con Fiftyone.
def createDataset(selectedClasses):

  # Limpia la salida.
  lblOutput.value = ''

  # Se deshabilitian los botones.
  btnCreateDataset.disabled = True
  btnExportDatasetToDrive.disabled = True

  # Filas seleccionadas.
  selectedRows = []

  # Se recorren todas las filas de widgets.
  for i in range(len(widgetRows)):

    # Comprueba si el usuario ha seleccionado la clase de la fila actual.
    if widgetRows[i][0].value:

      # Se añade la fila.
      selectedRows.append(i)

  # Si el usuario seleccionó al menos una fila.
  if len(selectedRows) > 0:

    # Borrado de clases antiguas.
    selectedClasses.clear()

    # Se borran todas las imágenes antiguas.
    if isdir('coco-2017'): rmtree('coco-2017')

    # Borrado de los Datasets anteriores.
    training = Dataset('training', overwrite=True)
    validation = Dataset('validation', overwrite=True)

    # Recorre las filas seleccionadas por el usuario.
    for i in selectedRows:

      # Se añade la clase seleccionada.
      selectedClasses.append(widgetRows[i][0].description)

      # Solo se mostrarán los LOG de ERROR.
      with LoggingLevel(ERROR):

        # Dataset auxiliar para obtener imágenes solo de una clase concreta.
        trainingAux = load_zoo_dataset(
            name='coco-2017',
            split='train',
            classes=widgetRows[i][0].description,
            only_matching=True,
            max_samples=widgetRows[i][1].value,
            shuffle=True,
            label_types=('detections', 'segmentations'),
            dataset_name='trainingAux',
            drop_existing_dataset=True
        )

        # Mezcla el dataset de entrenamiento con el auxiliar.
        training.merge_samples(trainingAux)

        # Dataset auxiliar para obtener imágenes solo de una clase concreta.
        validationAux = load_zoo_dataset(
            name='coco-2017',
            split='validation',
            classes=widgetRows[i][0].description,
            only_matching=True,
            max_samples=widgetRows[i][2].value,
            shuffle=True,
            label_types=('detections', 'segmentations'),
            dataset_name='validationAux',
            drop_existing_dataset=True
        )

        # Mezcla el dataset de validación con el auxiliar.
        validation.merge_samples(validationAux)

    # Borrado de los datasets auxiliares.
    delete_datasets('*Aux')

    # Se le informa al usuario de que los datasets se crearon correctamente.
    lblOutput.value = 'Successful creation of dataset'

    # Habilita el botón para la exportación de datasets.
    btnExportDatasetToDrive.disabled = False

  else:

    # Se le informa al usuario de que debe seleccionar al menos una clase.
    lblOutput.value = 'Please, select at least one class to create the dataset'

    # Si existían los datasets .
    if dataset_exists('training'):

      # Habilita el botón para la exportación de datasets.
      btnExportDatasetToDrive.disabled = False

  # Se habilitia el botón para la creación de datasets.
  btnCreateDataset.disabled = False
  
# Manejador del evento OnClick.
btnCreateDataset.on_click(lambda obj : createDataset(selectedClasses))

# Se añade el widget a la lista de widgets del FlexBox.
widgetList.append(btnCreateDataset)

# Botón para exportar el dataset a Google Drive.
btnExportDatasetToDrive = Button(
    description='Export To Google Drive',
    style=BUTTON_STYLE,
    layout=COMMON_LAYOUT
)

# Función para exportar el dataset a Google Drive.
def exportDatasetToDrive(selectedClasses):

  # Se vacía la salida de texto que ve el usuario.
  lblOutput.value = ''

  # El dataset ha de tener un nombre propio para poder exportarlo.
  if len(txtDatasetName.value) > 0:

    # Se deshabilitian los botones.
    btnCreateDataset.disabled = True
    btnExportDatasetToDrive.disabled = True

    # Ruta a la carpeta del dataset.
    dsFolder = join(wDirectory, txtDatasetName.value)

    # Si la carpeta existía se borra.
    if isdir(dsFolder): rmtree(dsFolder)
  
    # Crea la carpeta del dataset.
    makedirs(dsFolder)

    # Carga los datasets de entrenamiento y validación.
    training = load_dataset('training')
    validation = load_dataset('validation')

    # Solo se mostrarán los LOG de ERROR.
    with LoggingLevel(ERROR):

      # Exporta el conjunto de entrenamiento al formato COCO.
      training.export(
          dataset_type=COCODetectionDataset,
          export_dir=dsFolder,
          label_field='segmentations',
          data_path='training',
          labels_path='training.json'
      )

      # Exporta el conjunto de validación al formato COCO.
      validation.export(
          dataset_type=COCODetectionDataset,
          export_dir=dsFolder,
          label_field='segmentations',
          data_path='validation',
          labels_path='validation.json'
      )
    
    # Exporta el conjunto de entrenamiento y validación: tf-records.
    %sx python3 create_coco_tf_records.py \
    --train_image_dir={join(dsFolder, 'training')} \
    --val_image_dir={join(dsFolder, 'validation')} \
    --train_annotations_file={join(dsFolder, 'training.json')} \
    --val_annotations_file={join(dsFolder, 'validation.json')} \
    --output_dir={dsFolder}

    # Crea un fichero para escribir el mapa de correspondencias.
    with open(join(dsFolder, 'label_map.pbtxt'), 'w') as file:
      
      # Se recorren todas las clases y sus identificadores.
      for id, name in enumerate(selectedClasses, start=1):
        
        # Escribe los datos en el fichero.
        file.write(f"item {{\n\tid:{id}\n\tname: '{name}'\n}}\n\n")

    # Imágenes y detecciones de los conjuntos "training" y "validation".
    tImages = training.count('detections')
    vImages = validation.count('detections')
    tDetections = training.count('detections.detections')
    vDetections = validation.count('detections.detections')

    # Información básica a grabar en el fichero de resumen.
    dsInfo = {
      'name': txtDatasetName.value,
      'creation': datetime.now().strftime('%d/%m/%Y %H:%M:%S'),
      'num_classes': len(selectedClasses),
      'class_mames': selectedClasses,
      'total': {
         'images': tImages + vImages,
         'training_images': tImages,
         'validation_images': vImages,
         'detections': tDetections + vDetections,
         'training_detections': tDetections,
         'validation_detections': vDetections
      },
      'training': {},
      'validation': {}
    }

    # Información calculada para el fichero de "summary".
    for className in selectedClasses:

      # Vista de un sub-conjunto de imágenes para el entrenamiento.
      tView = training.filter_labels('detections', F('label') == className)
      
      # Vista de un sub-conjunto de imágenes para la validación.
      vView = validation.filter_labels('detections', F('label') == className)
      
      # Imágenes y detecciones de la clase actual en el conjunto "training".
      dsInfo['training'][className] = {
        'images': tView.count('detections'), 
        'detections': tView.count('detections.detections')    
      }

      # Imágenes y detecciones de la clase actual en el conjunto "validation".
      dsInfo['validation'][className] = {
        'images': vView.count('detections'), 
        'detections': vView.count('detections.detections')    
      }
    
    # Graba el contenido del diccionario en el fichero "summary.json"
    with open(join(dsFolder, 'summary.json'), 'w') as file: dump(dsInfo, file)

    # Si la carpeta ya existía en Drive, entonces se borra.
    if isdir(join(pathToDriveDatasets , txtDatasetName.value)):
      rmtree(join(pathToDriveDatasets, txtDatasetName.value))

    # Mueve la carpeta de Colab a la carpeta de anotaciones de Drive.
    move(
      join(wDirectory, txtDatasetName.value),
      join(pathToDriveDatasets, txtDatasetName.value)
    )

    # Se le indica al usuario que el dataset se exportó a Google Drive.
    lblOutput.value = 'Sucessful export of the last created dataset'

    # Se habilitian los botones.
    btnCreateDataset.disabled = False
    btnExportDatasetToDrive.disabled = False

  else:
    
    # Para exportar el dataset es necesario que su carpeta tenga nombre.
    lblOutput.value = 'Please, give a name to the dataset'

# Manejador del evento OnClick.
btnExportDatasetToDrive.on_click(lambda obj : exportDatasetToDrive(selectedClasses))

# Botón inicialmente deshabilitado.
btnExportDatasetToDrive.disabled = True

# Se añade el widget a la lista de widgets del FlexBox.
widgetList.append(btnExportDatasetToDrive)

# FlexBox para mostrar todos los widgets de forma vertical, salvo la salida.
fbPanel = Box(
    children=widgetList,
    layout=Layout(
        width='52%',
        display='flex',
        flex_flow='column',
        align_items='center',
        border='1px solid grey', 
        margin='20px 0px 20px 0px', 
        padding='10px 10px 10px 10px'
    )
)

# Label para proporcionar retroalimentación al usuario.
lblOutput = Label(
    value='',
    layout=Layout(width='52%')
)

# Muestra la GUI, compuesta por el panel FlexBox y el Label.
display(fbPanel, lblOutput)
```
